```python
import requests
import pandas as pd
from google.cloud import bigquery
import pandas_gbq
from google.colab import userdata


# 1. Get the Data
API_KEY = userdata.get('CoinmarketCap')
url = 'https://pro-api.coinmarketcap.com/v1/cryptocurrency/listings/latest'
parameters = {'start':'1', 'limit':'100', 'convert':'USD'}
headers = {
    'X-CMC_PRO_API_KEY': API_KEY,
    'Accepts': 'application/json'
}

response = requests.get(url, headers=headers)
data = response.json()['data']

# 2. Normalize "Messy" JSON
# Explain: CoinMarketCap sends nested data. You're flattening it for the warehouse.
df = pd.json_normalize(data)

# Clean column names for BigQuery compatibility
df.columns = df.columns.str.replace('.', '_', regex=False)
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    print(df)

# 3. Load to BigQuery (Free Tier)
#pandas_gbq.to_gbq(df, 'analytics-engineering-493415.production.raw_crypto_data',
 #         project_id='analytics-engineering-493415',
  #        if_exists='append')
```
